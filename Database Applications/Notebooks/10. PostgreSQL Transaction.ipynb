#### How to handle BEGIN, COMMIT and ROLLBACK statements;

A database transaction is a single unit of work that consists of one or more operations.

* A PostgreSQL transaction is atomic, consistant, isolated and durable.
* These properties are often referred to as ACID.
**Atomicity* -  It guarntees that the trasaction completes in an all or nothing manner
**Consistancy** - ensures that the change to data written to the database must be valid and follow predefined rules.
**Isolation**- determines how transaction integrity is visible to other transactions.
**Durablity** it make sure that transaction that have been committed will be stored in the database permanently.

#### To start a transaction, following statements can be used.

BEGIN TRANSACTION;or

BEGIN WORK or

BEGIN

#### Commit a transaction
COMMIT WORK; or

COMMIT TRANSACTION; or

COMMIT

After execution of COMMIT statement, PostgreSQL guarntees that the change will be durable if a crash happens.

#### Create a new table for Transaction demo

#### Create a new table for transaction purpose.

CREATE TABLE accounts (
    id INT GENERATED BY DEFAULT AS IDENTITY,
    name VARCHAR(100) NOT NULL,
    balance DEC(15,2) NOT NULL,
    PRIMARY KEY(id)
);

#### Insert two rows in the accounts table.
INSERT INTO accounts(name,balance)
VALUES('Bob',10000);

INSERT INTO accounts(name,balance)
VALUES('Alice',10000);

BEGIN and COMMIT example

#### start a transaction
BEGIN;

#### deduct 1000 from account 1
UPDATE accounts 
SET balance = balance - 1000
WHERE id = 1;

#### add 1000 to account 2
UPDATE accounts
SET balance = balance + 1000
WHERE id = 2; 

#### select the data from accounts
SELECT id, name, balance
FROM accounts;

#### commit the transaction
COMMIT;

Rolling back a transaction
To roll back or undo the change of the current transaction, use any of the following statement:
ROLLBACK TRANSACTION;
ROLLBACK WORK;
ROLLBACK;

#### begin the transaction
BEGIN;

#### deduct the amount from the account 1
UPDATE accounts 
SET balance = balance - 1500
WHERE id = 1;

#### add the amount from the account 3 (instead of 2)
UPDATE accounts
SET balance = balance + 1500
WHERE id = 3; 

#### Roll back the transaction
ROLLBACK;
