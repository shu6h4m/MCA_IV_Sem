```python
# import required libraries
import psycopg2
import pandas as pd
import pandas.io.sql as sqlio
```

```python
# Set up connection with NorthWind Database
#create database connection variable 
conn = psycopg2.connect(user="postgres", password="rohit", host="localhost", database="hotels")

conn.autocommit = True
cursor = conn.cursor()
```

```python
#execute query and save it to a variable
query = """
CREATE TABLE zomato
(
    "Restaurant ID" integer NOT NULL, "Restaurant Name" varchar, 
    "Country Code" integer NOT NULL, "City" varchar,
    "Address" text, "Locality" text, "Locality Verbose" text,
    "Longitude" double precision , "Latitude" double precision, "Cuisines" text,
    "Average Cost for two" integer, "Currency" char(50), 
    "Has Table booking" varchar, 
    "Has Online delivery" varchar ,
    "Is delivering now" varchar, 
    "Switch to order menu" varchar,
    "Price range" integer, "Aggregate rating" double precision , "Rating color" varchar,
    "Rating text" varchar, "Votes" integer
);
"""
cursor.execute(query)
```

```python
query = """
COPY zomato
(
    "Restaurant ID" , "Restaurant Name" , "Country Code" , "City" ,
    "Address", "Locality", "Locality Verbose" , "Longitude", "Latitude", 
    "Cuisines", "Average Cost for two", "Currency",  "Has Table booking",
    "Has Online delivery", "Is delivering now", "Switch to order menu",
    "Price range", "Aggregate rating", "Rating color", "Rating text",
    "Votes"
)
FROM 'D:\MCA\Practical\Database Application\zomato.csv'
DELIMITER ','
CSV HEADER
encoding 'windows-1251';
"""

cursor.execute(query)
```
