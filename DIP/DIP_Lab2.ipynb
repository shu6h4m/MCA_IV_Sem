```python
#Importing required libraries:
import PIL
from PIL import Image
 
import numpy as np
from skimage import io
#For ploting Histograph of image:
import matplotlib.pyplot as plt
```

```python
#opening and displaying image
fotu=Image.open('color.png') #Opening image and assigning it to 'fotu' variable
display(fotu)                #or fotu.show() can be used
```

```python
#Cheching image's details
print(fotu.format) #Image Format
print(fotu.size)   #Image Size
print(fotu.mode)   #Image Mode

#For image intensity
fotuSize= fotu.size
pixels = fotu.load()
```

```python
#Chnaging Intensity
for i in  range(0,(fotuSize[0]-1)):
   for j in range(0,(fotuSize[1]-1)):
      color = fotu.getpixel((i,j))

      redpixels = 255-color[0]
      greenpixels = 255-color[1]
      bluepixels = 255-color[2]
      fotu.putpixel((i,j),(redpixels,greenpixels,bluepixels))
display(fotu)
```

```python
#ploting the historgam for the given image
fotu2=io.imread('color.png')
ax=plt.hist(fotu2.ravel(),bins=256)
plt.show()
```

```python
r, b, g = fotu.split()
r.histogram()
b.histogram()
g.histogram()
```

```python
# Ploting Histogram for fotu2
_ = plt.hist(fotu2[:, :, 0].ravel(), bins = 8, color = 'red', alpha = 0.5)
_ = plt.hist(fotu2[:, :, 1].ravel(), bins = 8, color = 'Green', alpha = 0.5)
_ = plt.hist(fotu2[:, :, 2].ravel(), bins = 8, color = 'Blue', alpha = 0.5)
_ = plt.xlabel('Intensity Value')
_ = plt.ylabel('Count')
_ = plt.legend([ 'Red_Channel', 'Green_Channel', 'Blue_Channel'])
plt.show()
```
