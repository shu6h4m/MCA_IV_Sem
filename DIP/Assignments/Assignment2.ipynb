**Assignment 2 : To apply the following operations on the given images**

    1. Unsharp masking and Highboost filtering
    2. Laplacian filter

<hr>

```python
#Importing required libraries:

import PIL
import cv2
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
#opening and displaying given images using CV

fotu = cv2.imread('color.png')
cv2_imshow(fotu)

fotu2=cv2.imread('sample.png')
cv2_imshow(fotu2)
```

**1. Performing Unsharp masking and Highboost filtering**

```python
#Bluring the image and thn subtracting it from original

blurFotu = cv2.GaussianBlur(fotu, (1,1), 2.0)
unsharpFotu = cv2.addWeighted(fotu, 2.0, blurFotu, -1.0, 0)
cv2_imshow(unsharpFotu)
```

```python
#Bluring the image and thn subtracting it from original

blurFotu2 = cv2.GaussianBlur(fotu2, (1,1), 2.0)
unsharpFotu2 = cv2.addWeighted(fotu2, 2.0, blurFotu2, -1.0, 0)
cv2_imshow(unsharpFotu2)
```

**2. Applying Laplacian filter**

```python

fotu3 = cv2.imread('color.png', cv2.IMREAD_COLOR) # Loading image in greyscale

fotu3 = cv2.GaussianBlur(fotu3,(3,3),0) # Applying Gaussian Blur

fotuGray = cv2.cvtColor(fotu3, cv2.COLOR_BGR2GRAY) # Converting image to grayscale

# Now Applying Laplace Function
fotuLap = cv2.Laplacian(fotuGray, cv2.CV_16S, ksize=3)

# Converting back & displaying
fotuFinal = cv2.convertScaleAbs(fotuLap)
cv2_imshow(fotuFinal)
```
